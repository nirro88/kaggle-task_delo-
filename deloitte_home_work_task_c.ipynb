## Reduce the dimensionality 


### for each question positive or negative there is another opposite question for it
#### example: 
#### I am confident I will be able to make upcoming payments    and      I am concerned I will not be able to make upcoming paymentsÂ 

## The solution:
#### so will use it to reduce dimension and change the scores.
#### for example if a person has 5 in neg7 it will become 1 at column pos1, 
#### if he has 4 in neg7 it will become 2 at column pos1 and so on


```python
# replace the values in columns - ['neg7','pos8','neg8','pos9','neg9', 'pos10', 'neg10', 
#                                 'pos11', 'neg11', 'pos12', 'neg12', 'pos13','neg13']
# if value is 5 replace to 1
# if value is 1 replace to 5
# if value is 4 replace to 2
# if value is 2 replace to 4
# if value is 3 save value

d_replace = {5:1,1:5,4:2,2:4}
replace_value_columns_list = score_main_data.columns[16:29].to_list()
for col in replace_value_columns_list:
    score_main_data.replace({col:d_replace},inplace=True)
score_main_data.head(30)
```

### rank the country by continent

###  weights column
##### I believe there is a negative relationship between a person's financial well-being and weights column
#### we saw that in previos plot (income vs weights)
#### if a person comes from a 'developing country' they will get a score depending on their income

```python
# create a column for new weights
# if a person comes from a 'developing country' they will get a score depending on their income

def new_weight(row):
    list_high_diff = ['Brazil', 'China', 'Mexico','Italy', 'Poland', 'Republic of Korea', 'Spain']
    d_rank_income_for_weight = {'Low':10,'Med':5,'High':-3}
    if row['country'] in list_high_diff:
        if row['income'] == 'Low':
            val = row['weights'] * d_rank_income_for_weight['Low']
        elif row['income'] == 'Med':
            val = row['weights'] * d_rank_income_for_weight['Med']
        elif row['income'] == 'High':
            val = row['weights'] * d_rank_income_for_weight['High']
    else:
        val = row['weights']
    return val



score_main_data_appended['weights_new'] = -score_main_data_appended.apply(new_weight,axis=1)
```

## built the score 
### The method - the financial well-being score is created from the summation of all important columns

```python
sum_col_list = score_main_data_appended.drop(columns= ['gender','employment','country','age','income','weights']).columns.to_list()
# print(sum_col_list)
score_main_data_appended['final_score'] = score_main_data_appended[sum_col_list].sum(axis='columns')
score_main_data_appended.head(5)
```
